```python
#1. Write a Python function find_smallest_multiple(n: int) that uses a while loop to find the smallest positive integer x such that x is divisible by all numbers from 1 to n. Conditions: i. The function should keep incrementing x by 1 until it finds a number that satisfies the condition. ii. Your solution should handle the edge case where n = 1 efficiently, returning 1 directly since 1 is divisible by itself.
def find_smallest_multiple(n: int):
    if n == 1:
        return 1
    x = 1
    while True:
        divisible = True
        for i in range(1, n+1):
            if x % i != 0:
                divisible = False
                break
        if divisible:
            return x
        x += 1

# Example
n = 5
print("Smallest multiple:", find_smallest_multiple(n))
```

```python
#2. Write a function cubesum() that accepts an integer and returns the sum of the cubes of individual digits of that number. Use this function to make functions PrintArmstrong() and isArmstrong() to print Armstrong numbers and to find whether is an Armstrong number.
def cubesum(num):
    total = 0
    for digit in str(num):
        total += int(digit)**3
    return total
def isArmstrong(num):
    return num == cubesum(num)
def PrintArmstrong(start, end):
    for i in range(start, end+1):
        if isArmstrong(i):
            print(i, end=" ")
PrintArmstrong(100, 999)
```

```python
#4. Write a program to input a decimal number and print its equivalent binary, octal, and hexadecimal using operators.
num = int(input("Enter a decimal number: "))
binary = bin(num)[2:]      
octal = oct(num)[2:]       
hexa = hex(num)[2:]        
print("Binary:", binary)
print("Octal:", octal)
print("Hexadecimal:", hexa.upper())
```

```python
#6. Write a program that takes a sentence as input and counts the frequency of vowels, consonants, digits, and special characters separately
sentence = input("Enter a sentence: ")

vowels = consonants = digits = special = 0

for ch in sentence:
    if ch.isalpha():
        if ch.lower() in 'aeiou':
            vowels += 1
        else:
            consonants += 1
    elif ch.isdigit():
        digits += 1
    else:
        special += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special characters:", special)
```

```python
#7.Write a Python program to create a dictionary of students’ names as keys and their marks as values. Then: A. Print the student with the highest marks B. Print the student with the lowest marks
students = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 90,
    "David": 60
}
highest = max(students, key=students.get)
print("Highest Marks:", highest, students[highest])
lowest = min(students, key=students.get)
print("Lowest Marks:", lowest, students[lowest])
```
