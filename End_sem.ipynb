1. Given an integer N, write a Python program to:

a) Count the number of trailing zeros in the binary representation of N

b) Do not convert the number to a binary string

```python
def trailing_zeros(N):
  if N==0:
    return 0
  count=0
  while N%2==0:
    count+=1
    N//=2
  return count
N=4
print(trailing_zeros(N))
```

2. Write a program to compute F(n,r) where F(n,r) can be recursively defined as F(n,r)=F(n-1,r)+F(n-1, r-1)

```python
def F(n,r):
  if r==0 or r==n:
    return 1
  return F(n-1,r) + F(n-1,r-1)
print(F(5,2))
```
